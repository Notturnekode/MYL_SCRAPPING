```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import requests
```

```python
mainURL= 'https://myl.fandom.com/es/wiki/'
deckList = 'Categor%C3%ADa:Lista_de_Cartas'
```

```python
uClient = uReq(mainURL+deckList)
page_html = uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
```

```python
for table in page_soup.findAll("div"):
    print(table.get("class"))
```

```python
deckLT = page_soup.findAll("div", {"class": 'category-page__member-left'})
```
